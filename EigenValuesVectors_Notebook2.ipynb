Let the matrix be [[2,1,0] [1,3,-1] [0,-1,4]]

Write a piece of code to calculate the eigenvalue and eigenvector of matrix A. You may use Python functions to do calculate them. How can you verify that the eigenvectors that correspond to distinct eigenvalues are independent? Is there a way to code it in Python? Clue : Reduced Row Echelon.



#Solution Summary:


1. **Define Matrix A**: The problem involves computing eigenvalues and eigenvectors of a given matrix A and determining whether the eigenvectors corresponding to distinct eigenvalues are linearly independent.

2. **Calculate Eigenvalues and Eigenvectors**: Using NumPy's `np.linalg.eig()` function, we compute the eigenvalues and eigenvectors of matrix A.

3. **Check for Distinct Eigenvalues**: We check if the eigenvalues are distinct by comparing the number of unique eigenvalues with the total number of eigenvalues. If they are distinct, it indicates that each eigenvector corresponds to a different eigenvalue.

4. **Form Eigenvectors Matrix**: We form the eigenvectors matrix `M` by stacking the eigenvectors obtained from the calculation.

5. **Convert to SymPy Matrix**: The eigenvectors matrix `M` is converted to a SymPy Matrix `M_sympy` for symbolic computations.

6. **Reduce to RREF**: Using SymPy's `rref()` function, we reduce the eigenvectors matrix `M_sympy` to its Reduced Row Echelon Form (RREF).

7. **Check Linear Independence**: We count the number of pivot columns in the RREF and compare it with the number of distinct eigenvalues. If they are equal, the eigenvectors corresponding to distinct eigenvalues are linearly independent; otherwise, they are linearly dependent.



###Import Library:

```python
# Importing necessary libraries
import numpy as np  # For numerical computations
import sympy as sp  # For symbolic computations
```

###Define Matrix A:

```python
# Define the matrix A
A = np.array([[2, 1, 0],
              [1, 3, -1],
              [0, -1, 4]])
```

###Calculation of Eigen Values and Eigen Vectors:

```python
# Calculate eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

# Print eigenvalues and eigenvectors
print("Eigenvalues:")
for i in range(len(eigenvalues)):
    print("Eigenvalue {}: {:.4f}".format(i+1, eigenvalues[i]))

print("\nEigenvectors:")
for i in range(len(eigenvectors)):
    print("Eigenvector {}:".format(i+1), eigenvectors[:, i])
```

###To check whether Eigen Values are distinct:

```python
# Check if eigenvalues are distinct
if len(np.unique(eigenvalues)) == len(eigenvalues):
    print("\nEigenvalues are distinct.")
else:
    print("\nEigenvalues are not distinct.")
```

###Form Eigen Vector Matrix:

```python
# Form the eigenvectors matrix
M = np.column_stack(eigenvectors)
print("\nEigenvectors Matrix M:")
print(M)
```

###Convert to Sympy Matrix:

```python
# Convert M to a SymPy Matrix
M_sympy = sp.Matrix(M)
print(M_sympy)
```

###Reduce to RREF:

```python
# Reduce M to its Reduced Row Echelon Form (RREF)
RREF, pivot_columns = M_sympy.rref()
print("\nReduced Row Echelon Form (RREF) of M:")
print(RREF)
```

###Check Linear Independence:

```python
# Count the number of pivot columns
num_pivot_columns = len(pivot_columns)

# Check if the number of pivot columns is equal to the number of distinct eigenvalues
num_distinct_eigenvalues = len(np.unique(eigenvalues))

if num_pivot_columns == num_distinct_eigenvalues:
    print("\nThe eigenvectors corresponding to distinct eigenvalues are linearly independent.")
else:
    print("\nThe eigenvectors corresponding to distinct eigenvalues are linearly dependent.")
```

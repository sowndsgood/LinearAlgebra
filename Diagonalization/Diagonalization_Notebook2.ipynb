**P is invertible .Therefore columns of P are linearly independent.Why?**
If matrix  P is invertible, its columns are linearly independent due to the nonexistence of non-trivial linear combinations resulting in the zero vector.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A=np.array([[4,2],[1,5]])
print(A)
```

```python
P=np.array([[2,1],[-1,1]])
print(P)
```

```python
P_inv=np.array([[1/3,-1/3],[1/3,2/3]])
print(P_inv)
```

```python
D=np.array([[3,0],[0,6]])
print(D)
```

```python
b1=np.array([2,-1])
print(b1)
b2=np.array([1,1])
print(b2)
```

```python
x=np.array([2,1])
print(x)
```

```python
x_b_basis=np.dot(P_inv,x)
print(f"x in b basis is {x_b_basis}")
```

```python
x_b_basis_scaled=np.dot(D,x_b_basis)
print(f"x in b basis is {x_b_basis_scaled}")
```

```python
x_std_basis=np.dot(P,x_b_basis_scaled)
print(f"x in b basis is {x_std_basis}")
```

# Plot:

```python
#Section 1: Create Figure and Axis
# Create a new figure and axis
fig,ax=plt.subplots(figsize=(6,6))

#Section 2: Draw Original Basis Axes
# Draw the axes of the original basis
ax.plot([-10,20],[0,0],color='black',linestyle='-',linewidth=1,label='Standard Basis')
ax.plot([0,0],[-10,10],color='black',linestyle='-',linewidth=1)

#Section 3: Draw New Basis Axes
# Drawing the axes for new basis b
scaling_factor = 15
ax.plot([-scaling_factor * b1[0], scaling_factor * b1[0]], [-scaling_factor * b1[1], scaling_factor * b1[1]], color='yellow', linestyle='-', linewidth=1, label='Basis B')
ax.plot([-scaling_factor * b2[0], scaling_factor * b2[0]], [-scaling_factor * b2[1], scaling_factor * b2[1]], color='yellow', linestyle='-', linewidth=1)

#Section 4: Plot New Basis Vectors
# Plotting new basis vectors
ax.quiver(0,0,b1[0],b1[1],angles='xy',scale_units='xy',scale=1,color='green')
ax.quiver(0,0,b2[0],b2[1],angles='xy',scale_units='xy',scale=1,color='green')

#Section 5: Annotate Basis Vectors
# Annotate basis vectors with their coordinates
plt.text(b1[0]-1,b1[1]-0.5,'Basis b1 [2,-1]',fontsize=8,color='purple')
plt.text(b2[0]-5,b2[1]+0.1,'Basis b2 [1,1]',fontsize=8,color='purple')

#Section 6: Plot Points x and x_std_basis
# Plotting points x and x_std_basis
plt.plot(x[0], x[1],'bo',markersize=3)
plt.plot(x_std_basis[0],x_std_basis[1],'bo',markersize=3)

#Section 7: Plot Projections for x in New Basis b
# Plotting projections for x in new basis b
proj = b1 * x_b_basis[0]
ax.plot([proj[0], x[0]], [proj[1], x[1]], color='black', linestyle='--', linewidth=0.7)
proj = b2 * x_b_basis[1]
ax.plot([proj[0], x[0]], [proj[1], x[1]], color='black', linestyle='--', linewidth=0.7)

#Section 8: Plot Projections for Transformed x in New Basis b
# Plotting projections for transformed x in new basis b
proj = b1 * x_b_basis_scaled[0]
ax.plot([proj[0],x_std_basis[0]], [proj[1],x_std_basis[1]], color='black', linestyle='--', linewidth=0.7)
proj = b2 * x_b_basis_scaled[1]
ax.plot([proj[0],x_std_basis[0]], [proj[1],x_std_basis[1]], color='black', linestyle='--', linewidth=0.7)

#Section 9: Annotate Texts for x and x_transformed
# Annotating texts for x and x_transformed in original and basis
plt.text(x[0] + 0.7, x[1] + 0.1, 'x = [2,1] ; B-coords are [1/3,4/3]', fontsize=8)
plt.text(x_std_basis[0] + 0.7, x_std_basis[1] + 0.1, 'x_transformed = [10,7] ; B-coords : [1,8]', fontsize=8)

#Section 10: Set Axis Limits and Ticks
# Setting x and y axes limits
x_ticks = [-5, 0, 5, 10, 15, 20]
ax.set_xlim(-5, 20)
ax.set_ylim(-10, 10)
ax.set_xticks(x_ticks)
ax.set_yticks(np.arange(-10, 12, 2))

#Section 11: Show Grid, Legend, and Display Plot
# Showing the graph with dotted grid lines
plt.grid(linestyle='--', linewidth=0.9)
plt.legend()


pro1 = b1 * x_b_basis[0]
pro2 = b2 * x_b_basis[1]
pro3 = b1 * x_b_basis_scaled[0]
pro4 = b2 * x_b_basis_scaled[1]


# Coloring the area under x_transformed_std but excluding the area under x
x_poly = [pro2[0],pro4[0],x_std_basis[0],pro3[0],pro1[0],x[0]]
y_poly = [pro2[1],pro4[1],x_std_basis[1],pro3[1],pro1[1],x[1]]
ax.fill(x_poly, y_poly, color='mediumseagreen', alpha=0.5)
plt.show()
```

```python
import numpy as np  # Importing the NumPy library
import matplotlib.pyplot as plt  # Importing the pyplot module from the Matplotlib library
```

### What does T mean  here? (Clue : Try to interpret Dx)
In the expression T(x) = Dx, T represents a **Linear Transformation**. Here, D is a 2x2 matrix, and x is a vector in R^2. The operation "Dx" involves multiplying the vector x by the matrix D, effectively transforming x.

Given D is a diagonal matrix with diagonal entries 3 and 6, Dx scales the components of x independently. It scales the first component by 3 and the second component by 6.

So, T is a linear transformation that scales the input vector x in the x and y directions by factors of 3 and 6 respectively.

### What does basis vectors mean?
Basis vectors are fundamental vectors used to describe a vector space.

1.They are linearly independent, meaning that no vector in the set can be written as a linear combination of the others.

2.They span the vector space V, meaning that every vector in V can be expressed as a linear combination of the basis vectors.

###  D is what is called a Diagonal Matrix. Calculate D2 to understand why diagonal matrices can be useful.


```python
def matrix_multiplication(A, B):

    # Initialize the result matrix with zeros
    result = np.zeros((len(A), len(B[0])))

    # Iterate through rows of A
    for i in range(len(A)):
        # Iterate through columns of B
        for j in range(len(B[0])):
            # Iterate through rows of B
            for k in range(len(B)):
                # Perform the multiplication and accumulate the result
                result[i][j] += A[i][k] * B[k][j]

    return result
```

```python
# Define the diagonal matrix D using np.array
D = np.array([[3, 0],   # First row of the matrix
              [0, 6]])  # Second row of the matrix

# Print the matrix D
print(D)
```

```python
D_2 = matrix_multiplication(D, D)  # Computing D^2 using matrix multiplication
print(D_2)
```

```python
D_3 = matrix_multiplication(D_2, D)  # Computing D^3 using matrix multiplication
print(D_3)
```

Independent Scaling: In a diagonal matrix, each diagonal entry represents a scaling factor for its corresponding dimension. When multiplied with a vector, the diagonal matrix simply scales each component of the vector by the corresponding factor. This direct scaling operation simplifies the transformation process.

This demonstrates a key property of diagonal matrices: when multiplied by themselves, the resulting matrix remains diagonal, and each diagonal entry is squared. This property simplifies computations involving diagonal matrices, making them valuable in various mathematical and computational applications.

```python
e1 = np.array([1, 0])  # Define the vector e1 as [1, 0]
e2 = np.array([0, 1])  # Define the vector e2 as [0, 1]
```

```python
def matrix_vector_multiplication(A, v):

   # Check if matrix and vector dimensions are compatible for multiplication
    if A.shape[1] != v.shape[0]:
        raise ValueError("Matrix and vector dimensions are incompatible for multiplication.")

    # Initialize the result vector with zeros
    result = np.zeros(A.shape[0])

    # Perform matrix-vector multiplication
    for i in range(A.shape[0]):  # Iterate over rows of matrix A
        for j in range(A.shape[1]):  # Iterate over columns of matrix A
            result[i] += A[i, j] * v[j]  # Multiply corresponding elements and accumulate the result

    return result
```

```python
# Compute the transformation of e1 under the transformation represented by matrix D
e1_trans = matrix_vector_multiplication(D,e1)
print(e1_trans)
```

```python
# Compute the transformation of e2 under the transformation represented by matrix D
e2_trans = matrix_vector_multiplication(D, e2)
print(e2_trans)
```

```python
# Set the figure size
plt.figure(figsize=(6,6))  # Specify the figure size as 6x6 inches

# Plot all vectors
plt.plot([0, e1[0]], [0, e1[1]], color='r', label='e1')  # Plot e1
plt.plot([0, e2[0]], [0, e2[1]], color='g', label='e2')  # Plot e2
plt.plot([0, e1_trans[0]], [0, e1_trans[1]], color='b', label='e1_trans')  # Plot e1_trans
plt.plot([0, e2_trans[0]], [0, e2_trans[1]], color='y', label='e2_trans')  # Plot e2_trans

# Add arrowheads to all vectors
plt.arrow(0, 0, e1[0], e1[1], head_width=0.1, head_length=0.2, color='r')  # Arrow for e1
plt.arrow(0, 0, e2[0], e2[1], head_width=0.1, head_length=0.2, color='g')  # Arrow for e2
plt.arrow(0, 0, e1_trans[0], e1_trans[1], head_width=0.1, head_length=0.2, color='b')  # Arrow for e1_trans
plt.arrow(0, 0, e2_trans[0], e2_trans[1], head_width=0.1, head_length=0.2, color='y')  # Arrow for e2_trans

# Set labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Transformed Basis Vectors')

# Add legend
plt.legend()

# Add grid
plt.grid(True)

# Set aspect ratio to 'equal'
plt.gca().set_aspect('equal')

# Set x-axis and y-axis limits
plt.xlim(-0.5, 7)
plt.ylim(-0.5, 7)

# Add labels near the plot
plt.text(e1[0] + 0.1, e1[1] + 0.1, 'e1', fontsize=10, color='r')
plt.text(e2[0] + 0.1, e2[1] + 0.1, 'e2', fontsize=10, color='g')
plt.text(e1_trans[0] + 0.1, e1_trans[1] + 0.1, 'e1_trans', fontsize=10, color='b')
plt.text(e2_trans[0] + 0.1, e2_trans[1] + 0.1, 'e2_trans', fontsize=10, color='y')

# Show the plot
plt.show()
```
